```python
!pip install feedparser python-dateutil

import feedparser
from dateutil import parser as date_parser
import datetime

# ----------------------------
# 1. RSS FEEDS (AI + Startups)
# ----------------------------
feeds = [
    "https://openai.com/blog/rss.xml",
    "https://www.anthropic.com/news/rss.xml",
    "https://deepmind.google/rss.xml",
    "https://developer.nvidia.com/devblogs/feed",
    "https://www.producthunt.com/feed?category=artificial-intelligence",
    "https://arxiv.org/rss/cs.AI",
    "https://arxiv.org/rss/cs.LG",
]

# ----------------------------
# 2. Filter only today's articles
# ----------------------------
today = datetime.datetime.utcnow().date()   # match with RSS UTC timestamps

def is_today(entry_date_str):
    try:
        dt = date_parser.parse(entry_date_str)
        return dt.date() == today
    except:
        return False

# ----------------------------
# 3. Keyword filters
# ----------------------------
include_keywords = [
    "release", "launch", "update", "feature",
    "workflow", "use case", "gpt", "llm",
    "api", "integration", "beta", "model", "ai"
]

exclude_keywords = [
    "podcast", "interview", "discount", "pricing", "review"
]

# ----------------------------
# 4. Scrape + filter for TODAY only
# ----------------------------
filtered_items = []

for url in feeds:
    feed = feedparser.parse(url)

    for entry in feed.entries:
        published = entry.get("published", "") or entry.get("updated", "")

        # Filter by today's date
        if not is_today(published):
            continue

        title = entry.get("title", "").lower()

        # keyword inclusion
        if any(word in title for word in include_keywords):
            if not any(block in title for block in exclude_keywords):
                filtered_items.append({
                    "title": entry.title,
                    "link": entry.link,
                    "source": feed.feed.get("title", "Unknown Source"),
                    "published": published
                })

# ----------------------------
# 5. Build Google Docs Friendly HTML
# ----------------------------
filename = f"AI_Today_{today}.html"

html = f"""
<html><head><meta charset='UTF-8'><title>AI Daily Radar ‚Äì {today}</title></head>
<body><h2>üî• AI DAILY RADAR ‚Äì {today} üî•</h2>
<p><em>Only today's articles. Auto-filtered.</em></p><hr>
"""

if not filtered_items:
    html += "<p>No new AI releases today yet.</p>"

for item in filtered_items:
    html += f"""
    <p>
        <strong>Title:</strong>
        <a href="{item['link']}" target="_blank">{item['title']}</a><br>
        <strong>Source:</strong> {item['source']}<br>
        <strong>Published:</strong> {item['published']}<br>
    </p>
    <hr>
    """

html += "</body></html>"

with open(filename, "w", encoding="utf-8") as f:
    f.write(html)

filename
```
